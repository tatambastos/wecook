```python
from flask import Flask, request, jsonify
from sklearn.metrics.pairwise import cosine_similarity
import psycopg2

app = Flask(__name__)

def convertir_a_vector(datos):
    # Crear vector de características
    vector_columns = [0] * 4

    id_value2 = datos.get('id_value')
    name2 = datos.get('name')
    url2 = datos.get('url')
    description2 = datos.get('description')

    # Establecer los valores en el vector como 1 para las características presentes en la receta
    if id_value2:
        vector_columns[0] = 1
    if name2:
        vector_columns[1] = 1
    if url2:
        vector_columns[2] = 1
    if description2 and len(description2) > 0:  # Verificar si la descripción no está vacía
        vector_columns[3] = 1

    return vector_columns

@app.route('/recetas_favoritas', methods=['POST'])
def recibir_datos():
    datos = request.json
    user_id = datos.get('userId')

    # Establecer conexión a la base de datos
    conn = psycopg2.connect(database="cookwe", user="postgres", password="admin", host="localhost", port="5434")

    # Crear un cursor para ejecutar consultas SQL
    cursor = conn.cursor()

    # Ejecutar la consulta SQL para obtener las recetas favoritas del usuario
    cursor.execute("SELECT id_value, name, url, description FROM recetas_favoritas WHERE user_id = %s", (user_id,))

    # Obtener todas las recetas favoritas del usuario
    recetas_favoritas = cursor.fetchall()

    # Cerrar el cursor y la conexión a la base de datos
    cursor.close()
    conn.close()

    # Crear el vector de características para comparar con las recetas favoritas
    vector_caracteristicas = convertir_a_vector(datos)

    # Calcular la similitud de coseno entre el vector de características y las recetas favoritas
    similarity = cosine_similarity([vector_caracteristicas], [convertir_a_vector(receta) for receta in recetas_favoritas])

    # Obtener la similitud como un valor entre 0 y 1
    similarity_scores = similarity[0]

    # Definir el umbral de similitud
    umbral_similitud = 0.8  # Puedes ajustar este valor según tus necesidades

    # Obtener los índices de las recetas más similares que superan el umbral
    indices_recetas_similares = [indice for indice, score in enumerate(similarity_scores) if score > umbral_similitud]

    # Obtener las recetas similares
    recetas_similares = [recetas_favoritas[indice] for indice in indices_recetas_similares]

    # Crear una lista para almacenar las recetas recomendadas
    recomendaciones = []

    # Agregar las recetas recomendadas a la lista
    for receta in recetas_similares:
        recomendaciones.append({
            'id_value': receta[0],
            'name': receta[1],
            'url': receta[2],
            'description': receta[3]
        })

    # Devolver la respuesta JSON con las recetas recomendadas
    return jsonify({'message': 'Datos recibidos correctamente',
        'recetas_recomendadas': recomendaciones
    })
if __name__ == '__main__':
    app.run()

```
